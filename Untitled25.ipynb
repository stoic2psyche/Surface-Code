```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_state_city

# Create a 9-qubit circuit
qc = QuantumCircuit(9)

# Apply Hadamard to the first qubit
qc.h(0)

# Apply CNOTs from qubit 0 to all others
for i in range(1, 9):
    qc.cx(0, i)

# Get the final statevector directly from the circuit
statevector = Statevector.from_instruction(qc)

# Display the statevector
print(statevector)

# Plot the city view of the statevector
plot_state_city(statevector, title="9-Qubit GHZ State")
```

```python
from qiskit_aer.noise import NoiseModel, pauli_error
from qiskit_aer import AerSimulator

# Define a bit-flip error
bit_flip = pauli_error([('X', 0.1), ('I', 0.9)])  # 10% chance of X error

# Add it to a noise model
noise_model = NoiseModel()

noise_model.add_all_qubit_quantum_error(bit_flip, ['id'])  # apply after each gate

# Use simulator
sim = AerSimulator(noise_model=noise_model)
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit_aer.noise import NoiseModel, pauli_error
from qiskit.visualization import plot_histogram
from qiskit.quantum_info import Statevector
import matplotlib.pyplot as plt
from qiskit import transpile
```

```python
qc = QuantumCircuit(9, 9)

# Create |+⟩ state on qubit 0
qc.h(0)

# Entangle with rest (GHZ state)
for i in range(1, 9):
    qc.cx(0, i)

# Add barrier for clarity
qc.barrier()

qc.measure(range(9), range(9))
qc.draw('mpl')
```

```python
bit_flip_error = pauli_error([('X', 0.1), ('I', 0.9)])  # 10% bit-flip

noise_model = NoiseModel()

# Apply error to all qubits for identity instruction
noise_model.add_all_qubit_quantum_error(bit_flip_error, ['id'])
```

```python
simulator = AerSimulator(noise_model=noise_model)

# Transpile for simulator
compiled_circuit = transpile(qc, simulator)

# Run
result = simulator.run(compiled_circuit, shots=1024).result()

# Get counts
counts = result.get_counts()

# Plot
plot_histogram(counts)
```
